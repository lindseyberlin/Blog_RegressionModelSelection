### Imports

```python
# Data import and manipulation
import pandas as pd

# Math
import numpy as np

# Some statistics, including some checks for normality and skew
from scipy import stats
from scipy.stats import norm, skew

# Scikit-Learn for all of our models
from sklearn.linear_model import LinearRegression

# Visualization
import matplotlib.pyplot as plt
# Prettier visualizations
import seaborn as sns
# Sets the default style for seaborn, adding the nice darkgrid background
sns.set()
```

```python
# Data was already broken into training and testing data
train = pd.read_csv("house-prices-advanced-regression-techniques/train.csv",
                    index_col=0)
test = pd.read_csv("house-prices-advanced-regression-techniques/test.csv",
                   index_col=0)
train.head()
```

### Exploratory Data Analysis

```python
# Setting the SalePrice column as the target, since it's what we're trying to
# predict, and dropping it from our independent variables dataframe
target_train = train["SalePrice"]
train = train.drop(columns = "SalePrice")
```

```python
# Let's check in on our target variable - is it normally distributed?
plt.figure(figsize=(8,6))
# Plotting the distribution, setting fit=norm to use scipy to plot a normal
# distribution to compare our distribution to a standard normal one
sns.distplot(target_train, fit=norm)

plt.title("Distribution of Sale Prices")
plt.xlabel("Sale Price in Dollars")
plt.ylabel("Frequency")
plt.show()
```

```python
# Also plotting a Q-Q plot, to show skew
plt.figure(figsize=(8,6))
stats.probplot(target_train, plot=plt)
plt.show()
```

```python
# Alas, not normally distributed. Let's try log transforming it
# Using numpy's log + 1 because my instructor swears by it? Thx Bryan
target_train_log = np.log1p(target_train)
```

```python
# Plotting the two of them together
f, (ax1, ax2) = plt.subplots(1,2, figsize=(12,6))

sns.distplot(target_train_log, fit=norm, ax=ax1)
stats.probplot(target_train_log, plot=ax2)

ax1.set_title("Distribution of Sale Prices")

plt.show()
```

```python
# Much more normal, this we can work with. So let's move on to our features
train.info()
```

```python
# Let's use our intuition, that bigger houses will sell for more, and look
# at first only at LotArea
lot_area = train["LotArea"]

# Plotting our lot_area feature to see how the data is distributed
f, (ax1, ax2) = plt.subplots(1,2, figsize=(12,6))

sns.distplot(lot_area, fit=norm, ax=ax1)
stats.probplot(lot_area, plot=ax2)

ax1.set_title("Distribution of Lot Area")

plt.show()
```

```python
# It's overall kind of normally distributed, but there are some extremes
# Let's log-normalize this one as well
lot_area_log = np.log1p(lot_area)
```

```python
# Plotting our lot_area feature to see how the data is distributed
f, (ax1, ax2) = plt.subplots(1,2, figsize=(12,6))

sns.distplot(lot_area_log, fit=norm, ax=ax1)
stats.probplot(lot_area_log, plot=ax2)

ax1.set_title("Distribution of Lot Area")

plt.show()
```

```python
# That's a little bit better. So let's actually get started with models!
```

### Linear Regression

```python
# Before we do this - the linear regression model requires that the x and y
# be 2D arrays, and our data are currently 1D series. So:
lot_area_log = np.array(lot_area_log).reshape(-1,1)
target_train_log = np.array(target_train_log).reshape(-1,1)
```

```python
# Starting out linear regression model instance, from sklearn
linreg = LinearRegression()
linreg.fit(lot_area_log, target_train_log)
```
